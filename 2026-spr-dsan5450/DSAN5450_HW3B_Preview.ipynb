## Question 1.4: Deriving *Welfare Weights* from "Backward-Looking" Utility Calculations

Returning to our Utilitarian framework, with the above point that now we are directly comparing deaths to deaths, we can evaluate the **observed** decision to carry out the torture in an "ex post facto" manner, **deriving** the relative value-of-life weights that this policy reveals.

This is another case where it becoomes **crucial to explicitly differentiate** between **antecedents** and **consequents!** The study you will replicate in HW4, "Machine Learning What Policies Value", builds upon the notion we start exploring in this question: that if we **analyze a policy from the perpsective of social welfare**, there are **two "moving parts"** that we need to "pin down":

This type of analysis casts **observed policies** as "instantiations" of an implicational statement that looks like the statements you looked at in HW1:

$$
[\text{Ethical System } s] \implies [\text{Observed policy }p\text{ can be justified (by }s\text{)}]
$$

The difference between HW1 and this question (slash HW4), however, is that we **don't assume *knowledge* of the ethical system $s$** that policymakers are invoking to justify the observed policy $p$!

Instead, we carry out what's called a **counterfactual analysis**, which allows us to "back out" ethical values from $p$ in the following way:

1.  [Some antecedents] Assume a social-ethical decision which will affect $n$ individuals, and a decision-maker choosing values $\mathbf{x} = (x_1, \ldots, x_n)$, where varying choices of $x_i$ will result in varying effects on individual $i$.
    
    In this case the decision-maker is the US/Egyptian intelligence services and the choices $x_i \in \{0, 1\}$ are whether to apply torture to $i$. In HW4 the decision-maker will be the Mexican Ministry for Health and Education, and the choices $x_i$ will be how many pesos to allocate to household $i$
1.  Assume only that $s$ involves *some* aggregation of individual utilities to produce social welfare values $W(u_i) = \sum_{i}\omega_i u_i(x_i)$ (this is the Social Welfare Function<u>**al**</u>, different from the Social Welfare Function!). This setup can, in fact, accommodate many different ethical systems; see Roemer (2007) for Utilitarian and Rawlsian versions and Roemer (2019) for a Kantian version.
1.  The policy $p$, as mentioned above, is some **choice of allocation** $\mathbf{x} = (x_1, \ldots, x_n)$. The fact that we **can** observe these $x_i$ values (by collecting **data** on the implementation of a given policy!) means that we can **infer** the **welfare weights $\omega_i$** that the policy **must have** assigned to each individual $i$ under our assumptions.
